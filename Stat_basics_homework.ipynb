```python
horse_colic_select.value_counts('2_Age')
#Делаем предположение, что поскольку в описании должны быть два значения для столбца Age: 1 или 2, то 
#значение в данных = 9 - скорее опечатка, и если Age не равен 1 (взрослая лошадь), 
#то вместо 9 должно быть значение 2 (молодая лошадь).
#Посмотрим на данные для Age = 9, они обычные. Исходя из предположения об опечатке, заменим значение 9 на 2 - исправим данные.
#Делаем это уже в последней сохраненной таблице
```

```python
horse_colic_select_dropna1.sort_values(by='2_Age', ascending = False).head()
```
